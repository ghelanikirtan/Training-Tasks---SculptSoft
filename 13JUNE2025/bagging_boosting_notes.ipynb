# Bagging vs Boosting:

- Bagging & Boosting are ensemble methods (combining multiple ml models for overall performance improvements)

- **Ensemble** includes the training of multiple ml models independently. Their outputs/preds are aggregated either by voting, ave., weighting. [`sklearn.ensemble`]


## Bagging [Bootstrap Aggregating]:

- Best example of this method is RandomForestClassifier algorithm. Which uses Multiple Decision Trees and it train them individuallly and later the output is aggregated on the base of voting.

- This trains all the models parallely.


## Boosting:

- This kinda includes the sequential training methodology.
1. Initial Weights selection.
2. Train sequentially: start with weak learner, evaluate it's perfomance (regression-> mean of target, classification -> initial class probs. (log-odds)). Increase weights of missclassified instances.
3. Repeat this process of weights adjustment.
4. Aggregate (weighted aggregation -> Accurate learners more influence) the predictions of all weak learners for final output.
