```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
x = np.linspace(0,5, 15)
y = x ** 2
```

## Function Based Method:

```python
plt.plot(x,y)
plt.xlabel('X - label_name')
plt.ylabel('Y - label_name')
plt.title('Title')
```

```python
# Muliple Plots:

plt.subplot(1,2,1)
plt.plot(x, y, 'r-.') 

plt.subplot(1,2,2)
plt.plot(y, x, 'b*-')
```

## Object Based Plotting:

```python
# Figure creation
fig = plt.figure()

# axes addition to figure
axes = fig.add_axes([.1, .1, .8, .8]) # left, bottom, width, height

axes.plot(x, y, 'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
```

Multiple Figure Creations:

```python
# One plot inside the other
fig = plt.figure()

ax1 = fig.add_axes([.1,.1,.8,.8])
ax2 = fig.add_axes([.2,.5,.4,.3])

ax1.plot(x, y)
ax1.set_title('Parent')

ax2.plot(y, x, 'r')
ax2.set_title('Child')
ax2.set_xlabel('X-lbl-ax2')
ax2.set_xlabel('Y-lbl-ax2')
```

### Functional Programming with Matplotlib:

```python
x = np.linspace(0, 5, 15)
y = x**x
print(
    f"""x: {x}
y: {y}"""
)
```

```python
fig, axes = plt.subplots() #grabbing both figure & axes...

axes.plot(x, y)
```

```python
# Multiple plotting:

fig, axes = plt.subplots(nrows=2, ncols=3)
# plt.tight_layout()

for i in axes:
    for j in i: 
        j.plot(x,y)
plt.tight_layout()
```

```python
# Figure Size:

fig = plt.figure(figsize = (8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y)
```

Legend:

```python
fig, axes = plt.subplots(figsize= (5,5))
axes.plot(x, x**2, label = 'x-squared')
axes.plot(x, x**3, label = 'x-cubed')
axes.legend()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y, color = 'purple', lw =3.5, alpha = .5)
```

More Line Styles & Markers:

```python
fig = plt.figure(figsize= (12, 6))
axes = fig.add_axes([0,0,1,1])

# Line Styles:
axes.plot(x, x-1, color='brown', lw=2, linestyle='--')
axes.plot(x, x, color='brown', lw=2, linestyle='-.')
axes.plot(x, x+1, color='brown', lw=2, linestyle=':')
axes.plot(x, x+2, color='brown', lw=2, linestyle='-')

# Markers:
axes.plot(x, x+3, color='red', lw=0, ls='-', marker='o')
axes.plot(x, x+3, color='red', lw=0, ls='-', marker='+')
axes.plot(x, x+3, color='red', lw=0, ls='-', marker='*')
axes.plot(x, x+3, color='red', lw=0, ls='-', marker='1', markersize=10)
axes.plot(x, x+3, color='red', lw=0, ls='-', marker='2', markersize=15)

axes.plot(x, x+7, color='purple', lw=2, ls='-', marker='o', markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='blue')
```

### Plot Types:

```python
from random import sample
import math
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.scatter(data, [np.random.normal(0, std, 100) for std in range(1,4)])
```

2. Histograms:

```python
data = sample(range(1, 1000), 50)
plt.hist(data)
```

3. Boxplots:

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].boxplot(data, vert=True, patch_artist = True)
axes[1].boxplot(data)
```
