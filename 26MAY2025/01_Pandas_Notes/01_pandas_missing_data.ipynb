# Missing Data:

````python
import numpy as np
import pandas as pd
````

````python
d = {
    'A': [1, 2, np.nan],
    'B': [5, np.nan, np.nan],
    'C': [1,2,3]
}

df = pd.DataFrame(d)
df
````

Dropping rows/cols having null values:

````python
df.dropna() # drop rows (axis = 0)
````

````python
df.dropna(axis=1) # drop cols (axis = 1)
````

````python
# Dropping according to the threshold values:
df.dropna(thresh=2) # atleast 2 NaN values will be dropped
````

Replacing/Filling Missing Values:

````python
df.fillna(value='Fill Value')
````

````python
# Filling missing values with mean()

df.fillna(value = df.mean(), inplace=True)
df
````

Group by Method:

````python
data = {
    "Company": ["GOOG","GOOG", "MSFT", "MSFT", "FB", "FB"],
    "Person": ['Sam', 'Charlie', 'Ani', 'Vanessa', 'Clay', 'Sarah'],
    "Sales": [200,120,340,124,243,350]
}

df = pd.DataFrame(data)
df
````

````python
byCompany = df.groupby('Company')
byCompany
````

````python
byCompany.mean('Sales')
````

````python
byCompany.sum()
````

````python
# It will ignore if grping is not possible...
df.groupby('Person').mean('Sales')
````

````python
df.describe()
````

````python
df.T
````

````python
df.groupby('Company').describe().transpose()
````

DataFrames for Concatenation:

````python
df1 = pd.DataFrame(
    {
        'A': ['A0', 'A1', 'A2', 'A3'],
        'B': ['B0', 'B1', 'B2', 'B3'],
        'C': ['C0', 'C1', 'C2', 'C3'],
        'D': ['D0', 'D1', 'D2', 'D3']},
    index=[0, 1, 2, 3])


df2 = pd.DataFrame(
    {
        'A': ['A4', 'A5', 'A6', 'A7'],
        'B': ['B4', 'B5', 'B6', 'B7'],
        'C': ['C4', 'C5', 'C6', 'C7'],
        'D': ['D4', 'D5', 'D6', 'D7']},
    index=[4, 5, 6, 7]) 


df3 = pd.DataFrame(
    {
        'A': ['A8', 'A9', 'A10', 'A11'],
        'B': ['B8', 'B9', 'B10', 'B11'],
        'C': ['C8', 'C9', 'C10', 'C11'],
        'D': ['D8', 'D9', 'D10', 'D11']},
    index=[8, 9, 10, 11])
````

````python
df3
````

````python
# Concatenation: [dimensions should be same along the axis to be concatenated]

df_list = [df1, df2, df3]
pd.concat(df_list)
# here columns are equal as dfs are concatenated along the rows.
````

````python
# doing concatination along the columns...
pd.concat(df_list, axis=1)
````

````python
pd.concat(list(map(lambda n: n.T, df_list)), axis=1)
````

DataFrames: (For Merging)

````python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})    
````

````python
left
````

````python
right
````

- Syntax: 
```
pd.merge(left_df, right_df, how='inner/outer/left/right/cross', on='key')
```

````python
pd.merge(left, right, how='inner', on='key')
````

````python
left = pd.DataFrame(
    {
        'key1': ['K0', 'K0', 'K1', 'K2'],
        'key2': ['K0', 'K1', 'K0', 'K1'],
        'A': ['A0', 'A1', 'A2', 'A3'],
        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame(
    {
        'key1': ['K0', 'K1', 'K1', 'K2'],
        'key2': ['K0', 'K0', 'K0', 'K0'],
        'C': ['C0', 'C1', 'C2', 'C3'],
        'D': ['D0', 'D1', 'D2', 'D3']})
````

````python
left
````

````python
right
````

````python

pd.merge(left, right, on='key1')
````

````python
pd.merge(left, right, on=['key1', 'key2']) # default inner
````

````python
# outer join
pd.merge(left, right, how = 'outer', on=['key1', 'key2'])
````

In-built Join function DataFrames:

````python
left = pd.DataFrame(
    {
        'A': ['A0', 'A1', 'A2'],
        'B': ['B0', 'B1', 'B2']
    },
    index=['K0', 'K1', 'K2']) 

right = pd.DataFrame(
    {
        'C': ['C0', 'C2', 'C3'],
        'D': ['D0', 'D2', 'D3']},
    index=['K0', 'K2', 'K3'])  
````

````python
left.join(right)
````

````python
right.join(left)
````

````python
left.join(right, how='outer')
````
