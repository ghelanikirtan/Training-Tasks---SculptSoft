## Importing Packages and Loading sample dataset:

```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
data.head(5)
```

```python
# Seaborn sample plotting:::


plt.figure(figsize=(5,3))

sns.lineplot(
    x = 'sepal_length',
    y = 'sepal_width',
    data = data
)

# Styling:::
sns.set_theme(style='dark', context="paper")

# we use matplotlib's canvas and all to enable plotting...
plt.title("Title here...")
plt.show()
```

```python
# color palette:

pal1 = sns.color_palette()
sns.palplot(pal1)

pal2 = sns.color_palette('PiYG', 11)
sns.palplot(pal2)

# & many more...
```

Bar Plots:

```python
data = sns.load_dataset('tips')
data.head(3)
```

```python
sns.set_style('darkgrid')
sns.barplot(
    x = 'sex',
    y = 'total_bill',
    data = data,
    palette="PiYG"
)
```

Count Plot:

```python
sns.countplot(
    x = 'sex',
    data=data
)
```

Box Plot:

```python
sns.boxplot(
    x = 'day',
    y = 'total_bill',
    data = data,
    hue='smoker'
)
```

violin plot:

```python
sns.violinplot(
    x = 'day',
    y = 'total_bill',
    data = data,
    hue='sex'
)
```

Strip Plot:

```python
sns.stripplot(
    x = 'day',
    y = 'total_bill',
    data = data,
    hue='smoker',
    dodge=True
)
```

Correlation [HeatMap]:

```python
corr_data = data.drop(['time', 'day', 'sex', 'smoker'], axis=1)
corr_data = corr_data.corr()
corr_data
```

```python
sns.heatmap(corr_data, annot=True)
```

ClusterMaps:

```python
data = sns.load_dataset('flights')
data.head(4)
```

```python
cmap_data =  pd.pivot_table(values ='passengers', index ='month', 
                    columns ='year', data = data)
cmap_data
```

```python
sns.clustermap(cmap_data, annot=True, cmap='viridis', standard_scale=1) # 0-1 range
```
