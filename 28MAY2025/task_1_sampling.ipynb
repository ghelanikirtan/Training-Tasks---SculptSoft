## Imbalance Data Sampling Example:


- Kaggle dataset link ðŸ”—: [kaggle](https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud/data)

- This dataset is already preprocessed. But major issue is that the data is imbalanced.

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
# samplers import 
from imblearn.under_sampling import RandomUnderSampler, ClusterCentroids, NearMiss
from imblearn.over_sampling import RandomOverSampler, SMOTE, ADASYN
```

```python
# Loading the dataset related to credit card details...
df = pd.read_csv('data/creditcard.csv')
df.head(5)
```

```python
df.describe()
```

```python
df['Class'].value_counts()
```

- So here there are less number of classes with label 1 (Fraud Detected)...
- We will perform both kind of sampling here undersampling as well as oversampling...

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

### Under Sampling:

```python
samplers = {
    'Original_Data' : None,
    RandomUnderSampler.__name__ : RandomUnderSampler(random_state=0),
    ClusterCentroids.__name__ : ClusterCentroids(random_state=0),
    NearMiss.__name__ : NearMiss(version=3, n_neighbors=3)
}
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,10))

for ax, (title, sampler) in zip(axes.ravel(), samplers.items()):
    
    X_resampled, y_resampled = sampler.fit_resample(X,y) if sampler else (X,y)
    sns.scatterplot(
        ax = ax,
        x = X_resampled['Time'],
        y = X_resampled['V1'],
        hue = y_resampled,
    )
    ax.set_title(title)
    print(f"{title}: {Counter(y_resampled).items()} [Class Distribution]")
```

### Over Sampling:

```python
samplers = {
    'Original_Data' : None,
    RandomOverSampler.__name__ : RandomOverSampler(random_state=0).set_params(shrinkage=0.3),
    SMOTE.__name__ : SMOTE(random_state=0),
    ADASYN.__name__ : ADASYN(random_state=0)
}
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,10))

for ax, (title, sampler) in zip(axes.ravel(), samplers.items()):
    
    X_resampled, y_resampled = sampler.fit_resample(X,y) if sampler else (X,y)
    sns.scatterplot(
        ax=ax,
        x = X_resampled['Time'],
        y = X_resampled['V1'],
        hue = y_resampled,
    )
    ax.set_title(title)
    print(f"{title}: {Counter(y_resampled).items()} [Class Distribution]")
```
