# Linear Regression:

- Finding linear relationship between the input features & target. [straight line]

**Hypothesis Function:**
- Simple Linear Regression: $h(x) = y = \theta_0 + \theta_1 x_1$
- Multiple Linear Regression: $h(x) = y = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + ... + \theta_n x_n$


Here, $\theta_0$ is Intercept & $\theta_i$ are the slopes/weights associated...

**Cost Function:**

- MSE: $J(\theta) = \frac{1}{2n} \sum_{i=1}^{n}(y_i(pred) - y_i(actual))^2$

- Gradient of Loss: $J'(\theta) = \frac{1}{n} \sum_{i=1}^{n}{(y_i(pred) - y_i(actual)) \cdot x_i}$ 

- Gradient Descent [weight updation]: $\theta_j := \theta_{j} - \alpha \cdot {J'(\theta)} $

- Accuracy is calculated using mean absolute error.

```python
import numpy as np, pandas as pd
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
# Sample Data:
X, y = make_regression(n_samples=1000,
                       n_features=12,
                       noise=50,
                       random_state=42)
X.shape
```

```python
# Random Feature plotting to check the centrality of the data.
i = np.random.randint(0,11)
plt.hist(X.T[i], bins=30)
plt.show()
```

```python
# Splitting:
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=142)
```

Linear Regression model fit and eval:

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression() 
```

```python
lr_model.fit(X_train, y_train)
```

```python
# predictions;
y_pred = lr_model.predict(X_test)
```

```python
# scores:
scores = lr_model.score(X_test, y_test)
print(f"R2 Score: {scores}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


print(f"""
MAE: {mean_absolute_error(y_test, y_pred)}
RMSE: {mean_squared_error(y_test, y_pred)**.5}
R2 Score: {r2_score(y_test, y_pred)}
      """)
```

```python
X, y = make_regression(n_samples=100,
                       n_features=1,
                       n_informative=1,
                       random_state=30,
                       noise=5)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=48)
```

```python
slr = LinearRegression()
slr.fit(X_train, y_train)
```

```python
y_pred = slr.predict(X_test)
```

```python
# scores:
scores = slr.score(X_test, y_test)
print(f"R2 Score: {scores}")
```

```python
print(f"""
MAE: {mean_absolute_error(y_test, y_pred)}
RMSE: {mean_squared_error(y_test, y_pred)**.5}
R2 Score: {r2_score(y_test, y_pred)}""")
```

```python
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.plot(X_test, y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
